```python
import folium
import geopandas as gpd
from geojson import Point, Feature, FeatureCollection, dump
from shapely.geometry import Point, Polygon
import pandas as pd

long= -19.326
lat= -72.826

centerpoint = [long,lat]

xmin= long - 0.014438
ymin= lat - 0.014438
xmax= long + 0.014438
ymax= lat + 0.014438


features = []
polygon = Polygon([[xmin,
		  ymin],
		  [xmin,
		  ymax],
		  [xmax,
		  ymax],
		  [xmin,
		  ymax],
		  [xmin,
		  ymin]])
features.append(Feature(geometry=polygon, properties={"Area": "Drescher"}))
feature_collection = FeatureCollection(features)

with open('Drescher.geojson', 'w') as f:
   dump(feature_collection, f)
```

```python
# Set location
m = folium.Map([long, lat], zoom_start=10)
boundsdata = r'C:\Users\15714\Desktop\Final Project\EP Colonies\Drescher\Drescher.geojson'
folium.GeoJson(boundsdata).add_to(m)
```

```python
import sentinelsat
from sentinelsat import SentinelAPI, read_geojson, geojson_to_wkt
from datetime import date
api = SentinelAPI('theoreticalmarsupial', 'Splootking2022','https://scihub.copernicus.eu/dhus')

boundsdata = r'C:\Users\15714\Desktop\Final Project\EP Colonies\Drescher\Drescher.geojson'
footprint = geojson_to_wkt(read_geojson(boundsdata))

products = api.query(
    footprint,
    platformname = 'Sentinel-2',
    date = ('20210107', '20210108'),
    processinglevel = 'Level-2A',
    cloudcoverpercentage = (0,20),
    limit=1
)

# Now download them!!
api.download_all(products)
```
